```python
import pandas as pd
```

```python
df = pd.read_csv('data\salaries_by_college_major.csv')
df.sample(5)
```

```python
print(df.shape)
print(df.columns)
df.isna()
```

```python
clean_df = df.dropna()
clean_df.tail()
```

```python
clean_df['Starting Median Salary'].idxmax()
clean_df['Undergraduate Major'].loc[43]
```

```python
# Highest Mid-Career Median Salary
high_mid_idx = clean_df['Mid-Career Median Salary'].idxmax()
high_mid_major = clean_df['Undergraduate Major'].loc[high_mid_idx]
high_mid_salary = clean_df['Mid-Career Median Salary'].loc[high_mid_idx]
print(f'The major with the highest Mid-Career Median Salary is {high_mid_major} with a salary of {high_mid_salary}')
# print(clean_df['Undergraduate Major'].loc[8])
# print(clean_df['Mid-Career Median Salary'].loc[8])
```

```python
# Lowest Starting Median Salary
low_start_idx = clean_df['Starting Median Salary'].idxmin()
low_start_major = clean_df['Undergraduate Major'].loc[low_start_idx]
low_start_salary = clean_df['Starting Median Salary'].loc[low_start_idx]
print(f'The major with the lowest Starting Median Salary is {low_start_major} with a salary of {low_start_salary}')
# print(clean_df['Undergraduate Major'].loc[49])
# print(clean_df['Starting Median Salary'].loc[49])
```

```python
# Lowest Mid-Career Median Salary
low_mid_idx = clean_df['Mid-Career Median Salary'].idxmin()
low_mid_major = clean_df['Undergraduate Major'].loc[low_mid_idx]
low_mid_salary = clean_df['Mid-Career Median Salary'].loc[low_mid_idx]
print(f'The major with the lowest Mid-Career Median Salary is {low_mid_major} with a salary of {low_mid_salary}')
# print(clean_df['Undergraduate Major'].loc[18])
# print(clean_df['Mid-Career Median Salary'].loc[18])
```

```python
# Difference between Starting and Mid-Career Median Salary
clean_df.sort_values('Spread', ascending=False).head()
```

```python
pd.options.display.float_format = '{:,.2f}'.format
clean_df.groupby('Group').agg({'Starting Median Salary': ['mean', 'min', 'max'], 'Mid-Career Median Salary': ['mean', 'min', 'max']})
```
